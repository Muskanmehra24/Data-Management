```python
class Student:
    def __init__(student, name, roll_number, marks1, marks2):
        student.name = name
        student.roll_number = roll_number
        student.marks1 = marks1
        student.marks2 = marks2
        
```

```python
def accept():
    name = input("Enter Name: ")
    roll_number = int(input("Enter Roll No: "))
    marks1 = int(input("Enter marks1: "))
    marks2 = int(input("Enter marks2: "))
    #age = int(input("Enter age: "))

    Data = Student(name, roll_number, marks1, marks2)
    return Data
```

```python
def view(students):
    print(" Display Student Data:")
    for Data in students:
        print(f"Name: {Data.name}, Roll No: {Data.roll_number}, marks1: {Data.marks1}, marks2: {Data.marks2}")
```

```python
def search(students, search_value):
    result = [Data for Data in students if Data.roll_number == search_value]
    if result:
        print("Search Results:")
        for students in result:
            print(f"Name: {Data.name}, Roll No: {Data.roll_number}, marks1: {Data.marks1}, marks2: {Data.marks2}")
    else:
        print("No students found")
```

```python
def update(students, roll_number):
    for Data in students:
        if Data.roll_number == roll_number:
            name = input("Enter updated name: ")
            marks1 = int(input("Enter updated marks1: "))
            marks2 = int(input("Enter updated marks2: "))
            Data.name = name
            Data.marks1 = marks1
            Data.marks2 = marks2
            print("Student information updated successfully.")
            return
    print("No student found with this roll number.")
```

```python
def delete(students, roll_number):
    removed = False
    for Data in students:
        if Data.roll_number == roll_number:
            students.remove(Data)
            print(f"Student with roll number {roll_number} removed from the list.")
            removed = True
            break
    if not removed:
        print("No student found with the specified roll number.")

students = []
```

```python
while True: 
    print("Which function do you want to use?") 
    print("1. Accept")
    print("2. view") 
    print("3. Search")
    print("4. Update")
    print("5. Delete")
    print("6. Exit") 
    choice = int(input("Enter your choice (1-6): "))
    
    if choice == 1:
        Data = accept()
        students.append(Data)
        
    elif choice == 2:
        view(students)
            
    elif choice == 3:
        search_value = int(input("Enter the roll number to search: "))
        search(students, search_value)
                
    elif choice == 4:
        if not students:
            print("No students found.")
        else:
            roll_number = int(input("Enter the roll number of the student to update: "))
            update(students, roll_number)
                            
    elif choice == 5:
        roll_number = int(input("Enter the roll number of the student to delete: "))
        delete(students, roll_number)
                                
    elif choice == 6:
        break
    else:
        print("Invalid choice. Please try again.")
```
